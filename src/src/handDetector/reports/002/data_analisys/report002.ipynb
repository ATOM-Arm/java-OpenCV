```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
df = pd.read_csv("../batch_results.csv")
```

```python
df.head(15)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='fingers', hue='gesture', palette='Set1', bins=30, kde=True)
plt.title('Fingers vs Gesture')
plt.xlabel('Fingers')
plt.ylabel('Gesture')
plt.grid(True)
plt.legend(title='Gesture')
plt.show()
```
